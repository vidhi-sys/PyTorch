# Manipulating Tensors


1.    Addition
2.    Multiplications
3.    Divisions
4.    Matrix multiplications






```python
a=torch.tensor([1,2,3])
a=a+10
b=torch.tensor([4,5,6])
print(a+b)
a=a*10
a=a-10
print(a)
x=torch.tensor([1,2,3])
# Multiplication Element Wise
torch.mul(x,10)
torch.add(x,3)
```

```python
# Matrix Multiplication [dot Product]
tensor_matrix=torch.tensor([1,2,4])
print(tensor_matrix*tensor_matrix)
print(torch.matmul(tensor_matrix,tensor_matrix))
print(1*1+2*2+4*4)
value=0
for i in range (len(tensor_matrix)):
  value+=tensor_matrix[i]*tensor_matrix[i]
print(value)
```

https://medium.com/@lakshmanaraja/most-efficient-way-to-do-matrix-multiplication-python-vs-pytorch-deep-learning-6e038239dfa

```python
import time

# Manual loop for dot product
value = 0
start_time_manual = time.time()
for i in range(len(tensor_matrix)):
  value += tensor_matrix[i] * tensor_matrix[i]
end_time_manual = time.time()
print(f"Manual loop time: {end_time_manual - start_time_manual} seconds")

# torch.matmul for dot product
start_time_matmul = time.time()
torch.matmul(tensor_matrix, tensor_matrix)
end_time_matmul = time.time()
print(f"torch.matmul time: {end_time_matmul - start_time_matmul} seconds")
```
